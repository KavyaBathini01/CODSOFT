```python
#to design a contact book
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

def add_contact(contact_list, name, phone):
    contact = Contact(name, phone)
    contact_list.append(contact)
    print(f"Contact '{name}' added successfully!")            #when new contact is added

def display_contacts(contact_list):
    if not contact_list:
       print("Contact book is empty.")                        #display the contact in the contact book
    else:
        print("List of Contacts:")
        for i, contact in enumerate(contact_list, start=1):
            print(f"{i}. {contact.name} - {contact.phone}")

def search_contacts(contact_list, search_term):
    results = []
    for contact in contact_list:
        if search_term.lower() in contact.name.lower() or search_term in contact.phone:
            results.append(contact)
    return results

def update_contact(contact_list, old_name, new_name, new_phone):
    for contact in contact_list:
        if contact.name == old_name:
            contact.name = new_name
            contact.phone = new_phone
            print(f"Contact '{old_name}' updated successfully!")

def delete_contact(contact_list, name):
    for contact in contact_list:
        if contact.name == name:
            contact_list.remove(contact)
            print(f"Contact '{name}' deleted successfully!")

def main():
    contact_list = []
    while True:
        print("Contact Book Menu:")
        print("1. Add Contact")
        print("2. Display Contacts")
        print("3. Search Contacts")
        print("4. Update Contact")
        print("5. Delete Contact")
        print("6. Exit")

        choice = input("Enter your choice (1/2/3/4/5/6): ")

        if choice == "1":
            name = input("Enter contact name: ")
            phone = input("Enter contact phone number: ")
            add_contact(contact_list, name, phone)
        elif choice == "2":
            display_contacts(contact_list)
        elif choice == "3":
            search_term = input("Enter search term (name or phone): ")
            search_results = search_contacts(contact_list, search_term)
            if search_results:
                print("Search Results:")
                for i, result in enumerate(search_results, start=1):
                    print(f"{i}. {result.name} - {result.phone}")
            else:
                print("No matching contacts found.")
        elif choice == "4":
            old_name = input("Enter the name of the contact to update: ")
            new_name = input("Enter the new name: ")
            new_phone = input("Enter the new phone number: ")
            update_contact(contact_list, old_name, new_name, new_phone)
        elif choice == "5":
            name = input("Enter the name of the contact to delete: ")
            delete_contact(contact_list, name)
        elif choice == "6":
            print("Exiting Contact Book. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a valid option (1/2/3/4/5/6).")

if __name__ == "__main__":
    main()
```
